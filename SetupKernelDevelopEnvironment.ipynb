## Setup environment

```bash
python3 -m venv .venv
source .venv/bin/activate
pip install --upgrade pip
pip install uv
pip install jupyterlab

jupyter lab --notebook-dir ~
```

In JuypterLab open terminal and here:
```bash
cd ..
uv init --package jupyter-micropython-extended-kernel
cd jupyter-micropython-extended-kernel
```

```bash
uv add "pyserial>=3.4"
uv add "websocket-client>=0.44"
uv add "bleak>=0.22.3"
uv add "unsync>=1.4.0"
```

Test it:
```bash
uv run jupyter-micropython-extended-kernel
```
gives:
```bash:
Hello from jupyter-micropython-extended-kernel!
```
