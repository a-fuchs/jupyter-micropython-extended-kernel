```micropython
iCount = 0

def printIntSliderValue( iValue ):
    global iCount
    if iCount % 10 == 0:
        print()
        print( "Controller: ", end = " " )
        iCount = 0
    print( f"{iValue:>3}", end = " " )
    iCount += 1
```

```micropython
%%iPython
from IPython.display import display
from ipywidgets import widgets, interact, Layout, IntSlider


out = widgets.Output()               # A new output with name 'out' where 'print' should print in 

@out.capture(clear_output=False)     # Redirect all output of 'print' into 'out'
def sliderEventListener( iValue ):
    micropython_run( 'printIntSliderValue', iValue )

slider = IntSlider( min=0, max=127-4, step=1, layout=Layout(width='80%', height='80px') )
interact( sliderEventListener, iValue=slider )

#display( slider )  # Show slider
display( out )     # and then output of 'print'
```
