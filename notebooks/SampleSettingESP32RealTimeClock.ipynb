# Example: setting the Esp32 real time clock to current date and time

It is not good coding style, but kept simple to see what is done.

```micropython
%serialconnect
```

First define a function `setDateTime` which sets the RTC (real time clock) of the ESP:

```micropython
import machine

rtc = machine.RTC()
isDaylightSavingTime = False


def setDateTime( aTimeIPython ):
    # aTimeIPython = ( year, month, day, hour, minute, second, weekDay, yearDay, isDaylightSavingTime )
    global rtc, isDaylightSavingTime
    # Parameter for RTC.datetime: (year, month, day, 0, hour, minute, second, 0)
    rtc.datetime( (aTimeIPython[0], aTimeIPython[1], aTimeIPython[2], 0, aTimeIPython[3], aTimeIPython[4], aTimeIPython[5], 0 ) )
    isDaylightSavingTime = aTimeIPython[8] != 0
```

Then set the RTC on the ESP to the current date and time from the local PC:

```micropython
%%iPython
import time
micropython_run( "setDateTime", list( time.localtime() ) )
```
